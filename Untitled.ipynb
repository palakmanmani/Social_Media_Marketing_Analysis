```python
import pandas as pd

campaigns = pd.read_csv(r"C:\Users\HP\Downloads\social_media_marketing_dataset\campaigns.csv")

influencers = pd.read_csv(r"C:\Users\HP\Downloads\social_media_marketing_dataset\influencers.csv")

posts = pd.read_csv(r"C:\Users\HP\Downloads\social_media_marketing_dataset\posts.csv")
```

```python
campaigns['campaign_date'] = pd.to_datetime(campaigns['campaign_date'])
```

```python
posts['post_date'] = pd.to_datetime(posts['post_date'])
```

```python
campaigns.info()
```

```python
posts.info()
```

```python
campaigns['month'] = campaigns['campaign_date'].dt.month
```

```python
campaigns['year'] = campaigns['campaign_date'].dt.year
```

```python
posts['day_name'] = posts['post_date'].dt.day_name()
```

```python
campaigns['CTR'] = (campaigns['clicks'] / campaigns['impressions']) * 100
```

```python
campaigns['ROI'] = ((campaigns['revenue'] - campaigns['ad_spend']) / campaigns['ad_spend']) * 100
```

```python
posts['engagement_score'] = (
    posts['likes'] +
    posts['comments'] +
    posts['shares']
)
```

```python
campaigns.groupby('platform')['revenue'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

campaigns.groupby('platform')['revenue'] \
.sum() \
.plot(kind='bar')

plt.title("Revenue by Platform")
plt.xlabel("Platform")
plt.ylabel("Revenue")
plt.show()
```

```python
campaigns.groupby('month')['revenue'] \
.sum() \
.plot(kind='line', marker='o')

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
campaigns.groupby('platform')['ROI'] \
.mean() \
.plot(kind='bar')

plt.title("Average ROI by Platform")
plt.xlabel("Platform")
plt.ylabel("ROI")
plt.show()
```

```python
posts.groupby('content_type')['engagement_score'] \
.mean() \
.plot(kind='bar')

plt.title("Engagement by Content Type")
plt.xlabel("Content Type")
plt.ylabel("Engagement")
plt.show()
```

```python
campaigns.groupby('month')[['ad_spend','revenue']] \
.sum() \
.plot(kind='bar')

plt.title("Ad Spend vs Revenue")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```
